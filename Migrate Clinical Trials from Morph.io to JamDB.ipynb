```python
import requests
import json

morph_api_url = "https://api.morph.io/okfn/opentrials-clinicaltrailsgov-data/data.json"
morph_api_key = "qRunqlA1Fu9lJs3ApuK7"

local_jam = 'http://localhost:1212/v1/namespaces/OPENTRIALS/collections/data/documents'
auth_header = 'eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJleHAiOjE0NTMxNDY2OTIsInN1YiI6InRyYWNrZWQtU0hBUkV8dXNlcnMtc2NyYXBpIn0.S7qrO-g60yTIqoJz9JCDItTTHL1smrPg6p6VBlBANDU'
id_key_name = 'nct_id'

limit = 1000
offset = 0
error = False
total = 0
first = True

with open('./clinical-data.json', 'a') as f:
    f.write('[')
    while not error:
        r = requests.get(morph_api_url, params={
          'key': morph_api_key,
          'query': "select * from 'data' limit {} offset {}".format(limit, offset)
        })
        if not r.ok:
            error = True
            print('Got Error!')
        else:
            count = len(r.json())
            if count == 0:
                error = True
                print('Done.')
            else:
                total = total + count
                print('Got {} things for a total of {}.'.format(count, total))

                temp_data = r.json()
                for item in temp_data:
                    if not first:
                        f.write(',')
                    else:
                        first = False
                    f.write(json.dumps(item, indent=4))
                    data_dict = dict(data=dict(
                                                type="documents",
                                                id=item[id_key_name],
                                                attributes=item)
                                            )
                    jam_r = requests.post(local_jam,
                                          json=data_dict,
                                          headers=dict(Authorization=auth_header))
                    if not jam_r.ok:
                        print(jam_r.content)

                offset = offset + limit
    f.write(']')
    f.close()
```
